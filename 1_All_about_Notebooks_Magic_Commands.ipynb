# All About Magic Commands in Notebook 

## Important Magic Commands 

- %md - allows to write markdown text , design your notebook
- %python - switch python language in the code cell 
- %sql - switch code cell to sql code cell
- %sh - allows to run shell commands , linux command 
- %fs - allows to intract with DBFS (databricks file system )
- %scala - scala code
- %pip - install python packages 
- %run -call another notebook or python 
- %r - allow to run  R language
